- í´ëì¤ ë°ì´í° ì¸í¸ì í¼ì² ë°ì´í° ì¸í¸ë¥¼ ë¶ë¦¬í´ ë³ëì ë°ì´í° ì¸í¸ë¡ ë³ëë¡ ì ì¥


```python
cust_df['var3'].replace(-999999, 2, inplace=True)
cust_df.drop('ID', axis=1, inplace=True)

# í¼ì² ì¸í¸ì ë ì´ë¸ ì¸í¸ ë¶ë¦¬, ë ì´ë¸ ì¹¼ë¼ì DataFrameì ë§¨ ë§ì§ë§ì ìì¹í´ ì¹¼ë¼ ìì¹ -1ë¡ ë¶ë¦¬
X_features = cust_df.iloc[:, :-1]
y_labels = cust_df.iloc[:, -1]
print('í¼ì² ë°ì´í° shape:{0}'.format(X_features.shape))
```

```python
# ìë³¸ ë°ì´í° ì¸í¸ìì íìµ ë°ì´í° ì¸í¸ì íì¤í¸ ë°ì´í° ì¸í¸ ë¶ë¦¬
# ë¹ëì¹­í ë°ì´í° ì¸í¸ì´ë¯ë¡ í´ëì¤ì¸ Target ê° ë¶í¬ëê° íìµ ë°ì´í°ì íì¤í¸ ë°ì´í° ì¸í¸ì ëª¨ë ë¹ì·íê² ì¶ì¶ëëì§ íì¸
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels, test_size=0.2, random_state=0)

train_cnt = y_train.count()
test_cnt = y_test.count()
print('íìµ ì¸í¸ Shape:{0}, íì¤í¸ ì¸í¸ Shape:{1}'.format(X_train.shape, X_test.shape))

print('íìµ ì¸í¸ ë ì´ë¸ ê° ë¶í¬ ë¹ì¨')
print(y_train.value_counts()/train_cnt)
print('\n íì¤í¸ ì¸í¸ ë ì´ë¸ ê° ë¶í¬ ë¹ì¨')
print(y_test.value_counts()/test_cnt)
```

íìµ/íì¤í¸ ë°ì´í° ì¸í¸ ëª¨ë TARGETì ê°ì ë¶í¬ê° ìë³¸ê³¼ ì ì¬íê² ì ì²´ ë°ì´í°ì 4% ì ëì ë¶ë§ì¡± ê°(1)ì¼ë¡ ë§ë¤ì´ì§

```python
# X_train, y_trainì ë¤ì íìµ ê²ì¦ ë°ì´í° ì¸í¸ë¡ ë¶ë¦¬.
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=0)
```

#### XGBoost ëª¨ë¸ íìµê³¼ íì´í¼ íë¼ë¯¸í° íë

```python
# XGBoostì íìµ ëª¨ë¸ì ìì±íê³  ìì¸¡ ê²°ê³¼ë¥¼ ROC AUCë¡ íê°
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# n_estimatorsë 500ì¼ë¡, random stateë ìì  ìí ìë§ë¤ ëì¼ ìì¸¡ ê²°ê³¼ë¥¼ ìí´ ì¤ì 
xgb_clf = XGBClassifier(n_estimators=500, learning_rate=0.05, random_state=156)

# ì±ë¥ íê° ì§íë¥¼ aucë¡, ì¡°ê¸° ì¤ë¨ íë¼ë¯¸í°ë 100ì¼ë¡ ì¤ì íê³  íìµ ìí
xgb_clf.fit(X_tr, y_tr, early_stopping_rounds=100, eval_metric="auc", eval_set=[(X_tr,y_tr),(X_val,y_val)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1])
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

íì¤í¸ ë°ì´í° ì¸í¸ë¡ ìì¸¡ ì ROC AUCë ì½ 0.8429ì´ë¤.

```python
# HyperOptë¥¼ ì´ì©í´ ë² ì´ì§ì ìµì í ê¸°ë°ì¼ë¡ XGBoostì íì´í¼ íë¼ë¯¸í° íëì ìí
from hyperopt import hp

# max_depthë 5ìì 15ê¹ì§ 1 ê°ê²©ì¼ë¡, min_child_weightë 1ìì 6ê¹ì§ 1 ê°ê²©ì¼ë¡
# colsample_bytreeë 0.5ìì 0.95 ì¬ì´, learning_rateë 0.01ìì 0.2 ì¬ì´ ì ê· ë¶í¬ë ê°ì¼ë¡ ê²ì.
xgb_search_space = {'max_depth': hp.quniform('max_depth', 5, 15, 1),
                    'min_child_weight': hp.quniform('min_child_weight', 1, 6, 1),
                    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 0.95),
                    'learning_rate': hp.uniform('learning_rate', 0.01, 0.2)}
```

```python
# ëª©ì  í¨ì ë§ë¤ê¸°
# ëª©ì  í¨ìë 3 Fold êµì°¨ ê²ì¦ì ì´ì©í´ íê·  ROC-AUC ê°ì ë°ííë -1ì ê³±í´ì£¼ì´ ìµë ROC-AUC ê°ì´ ìµì ë°íê°ì´ ëê² í¨

from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score

# fmin()ìì í¸ì¶ ì search_space ê°ì¼ë¡ XGBClassifier êµì°¨ ê²ì¦ íìµ í -1*roc_auc íê·  ê°ì ë°í.
def objective_func(search_space):
  xgb_clf = XGBClassifier(n_estimators=100,max_depth=int(search_space['max_depth']),
                            min_child_weight=int(search_space['min_child_weight']),
                            colsample_bytree=search_space['colsample_bytree'],
                            learning_rate=search_space['learning_rate'])

  # 3ê° k-fold ë°©ìì¼ë¡ íê°ë roc_auc ì§íë¥¼ ë´ë list
  roc_auc_list = [ ]

  # 3ê° k-fold ë°©ì ì ì©
  kf = KFold(n_splits=3)
  # X_trainì ë¤ì íìµê³¼ ê²ì¦ì© ë°ì´í°ë¡ ë¶ë¦¬
  for tr_index, val_index in kf.split(X_train):
    # kf.split(X_train)ì¼ë¡ ì¶ì¶ë íìµê³¼ ê²ì¦ index ê°ì¼ë¡ íìµê³¼ ê²ì¦ ë°ì´í° ì¸í¸ ë¶ë¦¬
    X_tr, y_tr = X_train.iloc[tr_index], y_train.iloc[tr_index]
    X_val, y_val = X_train.iloc[val_index], y_train.iloc[val_index]

    # early stoppingì 30íë¡ ì¤ì íê³  ì¶ì¶ë íìµê³¼ ê²ì¦ ë°ì´í°ë¡ XGBClassifier íìµ ìí
    xgb_clf.fit(X_tr, y_tr, early_stopping_rounds=30, eval_metric="auc",
                eval_set=[(X_tr, y_tr), (X_val, y_val)])

    # 1ë¡ ìì¸¡í íë¥ ê° ì¶ì¶ í roc auc ê³ì°íê³  íê·  roc auc ê³ì°ì ìí´ listì ê²°ê³¼ê° ë´ì
    score = roc_auc_score(y_val, xgb_clf.predict_proba(X_val)[:,1])
    roc_auc_list.append(score)

  # 3ê° k-foldë¡ ê³ì°ë roc_auc ê°ì íê· ê°ì ë°ííë,
  # HyperOptë ëª©ì í¨ìì ìµìê°ì ìí ìë ¥ê°ì ì°¾ì¼ë¯ë¡ -1ì ê³±í ë¤ ë°í
  return -1*np.mean(roc_auc_list)
```

```python
# fmin() í¨ìë¥¼ í¸ì¶í´ max_eval=50íë§í¼ ë°ë³µíë©´ì ìµì ì íì´í¼ íë¼ë¯¸í°ë¥¼ ëì¶í´ë³´ì
from hyperopt import fmin, tpe, Trials

trials = Trials()

# fmin() í¨ìë¥¼ í¸ì¶
# max_evals ì§ì ë íìë§í¼ ë°ë³µ í ëª©ì í¨ìì ìµìê°ì ê°ì§ë ìµì  ìë ¥ê° ì¶ì¶
best = fmin(
    fn=objective_func,
    space=xgb_search_space,
    algo=tpe.suggest,
    max_evals=50,  # ìµë ë°ë³µ íìë¥¼ ì§ì í©ëë¤.
    trials=trials,
    rstate=np.random.default_rng(seed=30)
)

print('best:', best)
```

```python
# n_estimatorsë¥¼ 500ì¼ë¡ ì¦ê° í ìµì ì¼ë¡ ì°¾ì íì´í¼ íë¼ë¯¸í°ë¥¼ ê¸°ë°ì¼ë¡ íìµê³¼ ìì¸¡ ìí
xgb_clf = XGBClassifier(
    n_estimators=500,
    learning_rate=round(best['learning_rate'], 5),
    max_depth=int(best['max_depth']),
    min_child_weight=int(best['min_child_weight']),
    colsample_bytree=round(best['colsample_bytree'], 5),
    eval_metric='auc',
    early_stopping_rounds=100,
    random_state=156
)

# evaluation metricì aucë¡, early stoppingì 100ì¼ë¡ ì¤ì íê³  íìµ ìí
xgb_clf.fit(
    X_tr, y_tr,
    eval_set=[(X_tr, y_tr), (X_val, y_val)],
    verbose=False
)

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:, 1])
print('ROC AUC: {0:.4f}'.format(xgb_roc_score))
```

ROC-AUCê° ì´ì  ìì ì 0.8429ìì íì´í¼ íë¼ë¯¸í° íë ì´í 0.8451ë¡ ê°ì ë¨

```python
# íëë ëª¨ë¸ìì ê° í¼ì²ì ì¤ìëë¥¼ í¼ì² ì¤ìë ê·¸ëíë¡ ëíë´ê¸°
from xgboost import plot_importance
import matplotlib.pyplot as pit
%matplotlib inline

fig, ax = pit.subplots(1,1,figsize=(10, 8))
plot_importance(xgb_clf, ax=ax , max_num_features=20, height=0.4)
```

#### LightGBM ëª¨ë¸ íìµê³¼ íì´í¼ íë¼ë¯¸í° íë

```python
# ìì XGBoost ìì  ì½ëìì ë§ë¤ì´ì§ ë°ì´í° ì¸í¸ë¥¼ ê¸°ë°ì¼ë¡ LightGBMì¼ë¡ íìµì ìííê³  ROC-AUCë¥¼ ì¸¡ì í´ë³´ì
from lightgbm import LGBMClassifier, early_stopping # ìì 

lgbm_clf = LGBMClassifier(n_estimators=500)

eval_set = [(X_tr, y_tr), (X_val, y_val)]
lgbm_clf.fit(
    X_tr, y_tr,
    eval_metric="auc",
    eval_set=eval_set,
    callbacks=[early_stopping(stopping_rounds=100)] # ì¬ê¸°ì ì¡°ê¸° ì¢ë£ ì¤ì ì´ ë¤ì´ê°ëë¤.
) # ìì 
lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:, 1])
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```

```python
# HyperOPtë¥¼ ì´ì©í´ ë¤ìí íì´í¼ íë¼ë¯¸í°ì ëí íë ìí
lgbm_search_space = {'num_leaves': hp.quniform('num_leaves', 32, 64, 1),
                     'max_depth': hp.quniform('max_depth', 100, 160, 1),
                     'min_child_samples': hp.quniform('min_child_samples', 60, 100, 1),
                     'subsample': hp.uniform('subsample', 0.7, 1),
                     'learning_rate': hp.uniform('learning_rate', 0.01, 0.2)
}
```

```python
# ëª©ì  í¨ì ìì±. ìì¤ ì½ëë ì XGBoostì í¬ê² ë¤ë¥´ì§ ìì¼ë©° LGBMClassifier ê°ì²´ë¥¼ ìì±íë ë¶ë¶ì´ ë¬ë¼ì§
def objective_func(search_space):
  lgbm_clf = LGBMClassifier(n_estimators=100,
                            num_leaves=int(search_space['num_leaves']),
                            max_depth=int(search_space['max_depth']),
                            min_child_samples=int(search_space['min_child_samples']),
                            subsample=search_space['subsample'],
                            learning_rate=search_space['learning_rate'])
  # 3ê° k-fold ë°©ìì¼ë¡ íê°ë roc_auc ì§íë¥¼ ë´ë list
  roc_auc_list = []

  # 3ê° k-fold ë°©ì ì ì©
  kf = KFold(n_splits=3)
  # X_trainì ë¤ì íìµê³¼ ê²ì¦ì© ë°ì´í°ë¡ ë¶ë¦¬
  for tr_index, val_index in kf.split(X_train):
    # kf.split(X_train)ì¼ë¡ ì¶ì¶ë íìµê³¼ ê²ì¦ index ê°ì¼ë¡ íìµê³¼ ê²ì¦ ë°ì´í° ì¸í¸ ë¶ë¦¬
    X_tr, y_tr = X_train.iloc[tr_index], y_train.iloc[tr_index]
    X_val, y_val = X_train.iloc[val_index], y_train.iloc[val_index]

    # early stoppingì 30íë¡ ì¤ì íê³  ì¶ì¶ë íìµê³¼ ê²ì¦ ë°ì´í°ë¡ XGBClassifier íìµ ìí.
    lgbm_clf.fit(X_tr, y_tr,
                 eval_metric="auc",
                 eval_set=[(X_tr, y_tr), (X_val, y_val)],
                 callbacks=[early_stopping(stopping_rounds=30)])

    # 1ë¡ ìì¸¡í íë¥ ê° ì¶ì¶ í roc auc ê³ì°íê³  íê·  roc auc ê³ì°ì ìí´ listì ê²°ê³¼ê° ë´ì.
    score = roc_auc_score(y_val, lgbm_clf.predict_proba(X_val)[:,1])
    roc_auc_list.append(score)

  # 3ê° k-foldë¡ ê³ì°ë roc_auc ê°ì íê· ê°ì ë°ííë,
  # HyperOptë ëª©ì í¨ìì ìµìê°ì ìí ìë ¥ê°ì ì°¾ì¼ë¯ë¡ -1ì ê³±í ë¤ ë°í.
  return -1*np.mean(roc_auc_list)
```

```python
# fmin()ì í¸ì¶íì¬ ìµì  íì´í¼ íë¼ë¯¸í° ëì¶
from hyperopt import fmin, tpe, Trials

trials = Trials()

# fmin() í¨ìë¥¼ í¸ì¶. max_evals ì§ì ë íìë§í¼ ë°ë³µ í ëª©ì í¨ìì ìµìê°ì ê°ì§ë ìµì  ìë ¥ê° ì¶ì¶
best = fmin(fn=objective_func, space=lgbm_search_space, algo=tpe.suggest,
            max_evals=50, # ìµë ë°ë³µ íìë¥¼ ì§ì í©ëë¤.
            trials=trials, rstate=np.random.default_rng(seed=30))

print('best:', best)
```

```python
# ì´ë¤ íì´í¼ íë¼ë¯¸í°ë¥¼ ì´ì©íì¬ LightGBMì íìµ íì íì¤í¸ ë°ì´í° ì¸í¸ìì ROC-AUCë¥¼ íê°
from lightgbm import early_stopping
from lightgbm import LGBMClassifier # ìì 
from sklearn.metrics import roc_auc_score # ì

lgbm_clf = LGBMClassifier(n_estimators=500, num_leaves=int(best['num_leaves']),
                          max_depth=int(best['max_depth']),
                          min_child_samples=int(best['min_child_samples']),
                          subsample=round(best['subsample'], 5),
                          learning_rate=round(best['learning_rate'], 5)
)

# evaluation metricì aucë¡, early stoppingì 100ì¼ë¡ ì¤ì íê³  íìµ ìí.
lgbm_clf.fit(X_tr, y_tr,
             eval_metric="auc",
             eval_set=[(X_tr, y_tr), (X_val, y_val)],
             callbacks=[early_stopping(stopping_rounds=100)]) # ìì 

lgbm_roc_score = roc_auc_score(y_test,lgbm_clf.predict_proba(X_test)[:,1])
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```

2. ì¤ë² ìíë§ : ì´ì ë°ì´í°ì ê°ì´ ì ì ë°ì´í° ì¸í¸ë¥¼ ì¦ìíì¬ íìµì ìí ì¶©ë¶í ë°ì´í°ë¥¼ íë³´íë ë°©ë². ìë³¸ ë°ì´í°ì í¼ì² ê°ë¤ì ìì£¼ ì½ê°ë§ ë³ê²½íì¬ ì¦ìí¨.

- Amount í¼ì²ë ì ì©ì¹´ë í¸ëì­ì ê¸ì¡ì ìë¯¸íë©°, Classë ë ì´ë¸ë¡ì 0ì ê²½ì° ì ì, 1ì ê²½ì° ì¬ê¸° í¸ëì­ìì´ë¤.

```python
from sklearn.model_selection import train_test_split

# ì¸ìë¡ ìë ¥ë°ì DataFrameì ë³µì¬í ë¤ Time ì¹¼ë¼ë§ ì­ì íê³  ë³µì¬ë DataFrame ë°í
def get_preprocessed_df(df=None):
  df_copy = df.copy()
  df_copy.drop('Time', axis=1, inplace=True)
  return df_copy
```

```python
# ì¬ì  ë°ì´í° ê°ê³µ í íìµê³¼ íì¤í¸ ë°ì´í° ì¸í¸ë¥¼ ë°ííë í¨ì.
def get_train_test_dataset(df=None):
  # ì¸ìë¡ ìë ¥ë DataFrameì ì¬ì  ë°ì´í° ê°ê³µì´ ìë£ë ë³µì¬ DataFrame ë°í
  df_copy = get_preprocessed_df(df)
  # DataFrameì ë§¨ ë§ì§ë§ ì¹¼ë¼ì´ ë ì´ë¸, ëë¨¸ì§ë í¼ì²ë¤
  X_features = df_copy.iloc[:,:-1]
  y_target = df_copy.iloc[:,-1]
  # train_test_split()ì¼ë¡ íìµê³¼ íì¤í¸ ë°ì´í° ë¶í . stratify=y_targetì¼ë¡ Stratified ê¸°ë° ë¶í 
  X_train, X_test, y_train, y_test = \
  train_test_split(X_features, y_target, test_size=0.3, random_state=0, stratify=y_target)
  # íìµê³¼ íì¤í¸ ë°ì´í° ì¸í¸ ë°í
  return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = get_train_test_dataset(card_df)
```

```python
print('íìµ ë°ì´í° ë ì´ë¸ ê° ë¹ì¨')
print(y_train.value_counts()/y_train.shape[0]*100)
print('íì¤í¸ ë°ì´í° ë ì´ë¸ ê° ë¹ì¨')
print(y_test.value_counts()/y_test.shape[0]*100)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def get_clf_eval(y_test, pred=None, pred_proba=None):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)
    # ROC-AUC ì¶ê°
    roc_auc = roc_auc_score(y_test, pred_proba)
    print('ì¤ì°¨ íë ¬')
    print(confusion)
    # ROC-AUC print ì¶ê°
    print('ì íë: {0:.4f}, ì ë°ë: {1:.4f}, ì¬íì¨: {2:.4f}, F1ï¼ {3:.4f}, AUCï¼{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
# ë¡ì§ì¤í± íê·ì LightGBM ê¸°ë°ì ëª¨ë¸ì´ ë°ì´í° ê°ê³µì ìííë©´ì ìì¸¡ ì±ë¥ì´ ì´ë»ê² ë³íëì§ ì´í´ë³´ê¸°
from sklearn.linear_model import LogisticRegression

Ir_clf = LogisticRegression(max_iter=1000)
Ir_clf.fit(X_train, y_train)
Ir_pred = Ir_clf.predict(X_test)
Ir_pred_proba = Ir_clf.predict_proba(X_test)[:,1]

# 3ì¥ìì ì¬ì©í get_clf_eval() í¨ìë¥¼ ì´ì©í´ íê° ìí.
get_clf_eval(y_test, Ir_pred, Ir_pred_proba)
```

```python
# ì¸ìë¡ ì¬ì´í·ë°ì Estimator ê°ì²´ì íìµ/íì¤í¸ ë°ì´í° ì¸í¸ë¥¼ ìë ¥ë°ìì íìµ/ìì¸¡/íê° ìí.
def get_model_train_eval(model, ftr_train=None, ftr_test=None, tgt_train=None, tgt_test=None):
  model.fit(ftr_train, tgt_train)
  pred = model.predict(ftr_test)
  pred_proba = model.predict_proba(ftr_test)[:,1]
  get_clf_eval(tgt_test, pred, pred_proba)
```

```python
# StandardScalerê° ìëë¼ ë¡ê·¸ ë³íì ìí
# ë¡ê·¸ ë³í : ë°ì´í° ë¶í¬ëê° ì¬íê² ìê³¡ëì´ ìì ê²½ì° ì ì©íë ì¤ì ê¸°ë² ì¤ íë

def get_preprocessed_df(df=None):
  df_copy = df.copy()
  # ëíì´ì log1p()ë¥¼ ì´ì©í´ Amountë¥¼ ë¡ê·¸ ë³í
  amount_n = np.log1p(df_copy['Amount'])
  df_copy.insert(0, 'Amount_Scaled', amount_n)
  df_copy.drop(['Time', 'Amount'], axis=1, inplace=True)
  return df_copy
```

ìê´ê´ê³ íí¸ë§µìì ë§¨ ìëì ìì¹í ê²°ì  ë ì´ë¸ì¸ Class í¼ì²ì ìì ìê´ê´ê³ê° ê°ì¥ ëì í¼ì²ë V14ì V17ì´ë¤. ì´ ì¤ V14ì ëí´ìë§ ì´ìì¹ë¥¼ ì°¾ìì ì ê±°í´ ë³´ê² ë¤.

```python
import numpy as np

def get_outlier(df=None, column=None, weight=1.5):
  # fraudì í´ë¹íë column ë°ì´í°ë§ ì¶ì¶, 1/4 ë¶ìì 3/4 ë¶ì ì§ì ì np.percentileë¡ êµ¬í¨.
  fraud = df[df['Class']==1][column]
  quantile_25 = np.percentile(fraud.values, 25)
  quantile_75 = np.percentile(fraud.values, 75)

  # IQRì êµ¬íê³ , IQRì 1.5ë¥¼ ê³±í´ ìµëê°ê³¼ ìµìê° ì§ì  êµ¬í¨.
  iqr = quantile_75 - quantile_25
  iqr_weight = iqr * weight
  lowest_val = quantile_25 - iqr_weight
  highest_val = quantile_75 + iqr_weight

  # ìµëê°ë³´ë¤ í¬ê±°ë, ìµìê°ë³´ë¤ ìì ê°ì ì´ìì¹ ë°ì´í°ë¡ ì¤ì íê³  DataFrame index ë°í.
  outlier_index = fraud[(fraud < lowest_val) | (fraud > highest_val)].index
  return outlier_index
```

```python
# get_outlier() í¨ìë¥¼ ì´ì©í´ V14ì¹¼ë¼ìì ì´ìì¹ ë°ì´í° ì°¾ê¸°
outlier_index = get_outlier(df=card_df, column='V14', weight=1.5)
print('ì´ìì¹ ë°ì´í° ì¸ë±ì¤:', outlier_index)
```

```python
# get_processed_df()ë¥¼ ë¡ê·¸ ë³í í V14 í¼ì²ì ì´ìì¹ ë°ì´í°ë¥¼ ì­ì íë ë¡ì§ì¼ë¡ ë³ê²½.
def get_preprocessed_df(df=None):
  df_copy = df.copy()
  amount_n = np.log1p(df_copy['Amount'])
  df_copy.insert(0, 'Amount_Scaled', amount_n)
  df_copy.drop(['Time', 'Amount'], axis=1, inplace=True)
  # ì´ìì¹ ë°ì´í° ì­ì íë ë¡ì§ ì¶ê°
  outlier_index = get_outlier(df=df_copy, column='V14', weight=1.5)
  df_copy.drop(outlier_index, axis=0, inplace=True)
  return df_copy

X_train, X_test, y_train, y_test = get_train_test_dataset(card_df)
print('### ë¡ì§ì¤í± íê· ìì¸¡ ì±ë¥ ###')
get_model_train_eval(lr_clf, ftr_train=X_train, ftr_test=X_test, tgt_train=y_train,
                     tgt_test=y_test)
print('### LightGBM ìì¸¡ ì±ë¥ ###')
get_model_train_eval(lgbm_clf, ftr_train=X_train, ftr_test=X_test, tgt_train=y_train,
                     tgt_test=y_test)
```

```python
# ì ìì ìì ìì±í íìµ í¼ì²/ë ì´ë¸ ë°ì´í°ë¥¼ SMOTE ê°ì²´ì fit_resample() ë©ìëë¥¼ ì´ì©í´ ì¦ìí ë¤ ë°ì´í°ë¥¼ ì¦ì ì ê³¼ ë¹êµ
# íìµ ë°ì´í° ì¸í¸ë§ ì¤ë² ìíë§ í´ì¼ í¨
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=0)
X_train_over, y_train_over = smote.fit_resample(X_train, y_train)
print('SMOTE ì ì© ì  íìµì© í¼ì²/ë ì´ë¸ ë°ì´í° ì¸í¸: ', X_train.shape, y_train.shape)
print('SMOTE ì ì© í íìµì© í¼ì²/ë ì´ë¸ ë°ì´í° ì¸í¸: ', X_train_over.shape, y_train_over.shape)
print('SMOTE ì ì© í ë ì´ë¸ ê° ë¶í¬: \n', pd.Series(y_train_over).value_counts())
```

```python
from sklearn.metrics import precision_recall_curve

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline

def precision_recall_curve_plot(y_test, pred_proba_c1):
  # threshold ndarrayì ì´ thresholdì ë°ë¥¸ ì ë°ë, ì¬íì¨ ndarray ì¶ì¶.
  precisions, recalls, thresholds = precision_recall_curve(y_test, pred_proba_c1)

  # Xì¶ì thresholdê°ì¼ë¡, Yì¶ì ì ë°ë, ì¬íì¨ ê°ì¼ë¡ ê°ê° Plot ìí. ì ë°ëë ì ì ì¼ë¡ íì
  plt.figure(figsize=(8,6))
  threshold_boundary = thresholds.shape[0]
  plt.plot(thresholds, precisions[0:threshold_boundary], linestyle='--', label='precision')
  plt.plot(thresholds, recalls[0:threshold_boundary], label='recall')

  # threshold ê° Xì¶ì Scaleì 0.1 ë¨ìë¡ ë³ê²½
  start, end = plt.xlim()
  plt.xticks(np.round(np.arange(start, end, 0.1), 2))

  # xì¶, yì¶ labelê³¼ legend, ê·¸ë¦¬ê³  grid ì¤ì 
  plt.xlabel('Threshold value'); plt.ylabel('Precision and Recall value')
  plt.legend(); plt.grid()
  plt.show()
```
